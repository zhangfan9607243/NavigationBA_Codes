# Topic 2 - SQL 基础查询

本章将介绍 SQL 查询的基础知识，包括基础的 `SELECT` 与 `FROM` 语句、`WHERE` 条件过滤、`ORDER BY` 排序以及 `TOP` 限制结果数量等内容。通过实际案例和练习，同学们将能够编写基本的 SQL 查询来从数据库中提取所需的数据。我们将逐步引导同学们理解 SQL 查询的结构和语法，帮助大家建立扎实的 SQL 基础，为后续章节的高级查询打下坚实的基础。

- **Topic 2.1 - SELECT 与 FROM 语句**：学习如何使用 `SELECT` 语句选择需要的列，以及使用 `FROM` 语句指定数据来源的表，并且学习使用 `DISTINCT` 语句去重查询。
- **Topic 2.2 - WHERE 条件过滤**：学习如何使用 `WHERE` 语句进行条件过滤，以及 SQL 中的比较运算和逻辑运算，以便从数据库中筛选出满足特定条件的数据。
- **Topic 2.3 - ORDER BY 语句数据排序与 TOP 结果限制**：学习如何使用 `ORDER BY` 语句对查询结果进行排序，掌握升序和降序排序的使用方法，以便更好地组织和展示查询结果，并且学习使用 `TOP` 语句限制查询结果的数量，以便更高效地获取所需的数据。
- **Topic 2.4 - SELECT 语句中的进阶数据运算**：学习 SQL 中三种主要的数据类型：数值型、字符串型和日期时间型，并且学习如何使用 SQL 进行数据计算和转换，包括使用函数进行数值计算、字符串操作和日期时间处理等，以便更高效地分析和处理数据。
- **Topic 2.5 - GROUP BY 语句与数据聚合运算**：学习如何使用 `GROUP BY` 语句对查询结果进行分组，并且学习 SQL 中的聚合函数（如 `COUNT`、`SUM`、`AVG`、`MAX`、`MIN`）来对分组后的数据进行统计和分析，以便使用 `HAVING` 语句对分组后的数据进行条件过滤，从而更深入地分析和理解数据。
