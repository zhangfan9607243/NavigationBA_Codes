## 2. SQL 中的表拼接

在 SQL 中，表拼接的操作使用 `UNION ALL` 关键字来实现：

```sql
SELECT column1, column2, ...
FROM table1
...

UNION ALL

SELECT column1, column2, ...
FROM table2
...
```

这个语法方式有些特殊，同学们之前可能没有见过，下面我们来逐步分析一下：

- 这段 SQL 语句由两段 `SELECT` 语句组成，每段 `SELECT` 语句都产生一个小表
- 按照我们上面的例子，第一段 `SELECT` 语句产生了张三李四那个学生表，第二段 `SELECT` 语句产生了 Alice Bob 那个学生表
- 这两段 `SELECT` 语句之间使用了 `UNION ALL` 关键字，这个关键字的作用就是将两段 `SELECT` 语句产生的小表按照行的方向拼接在一起，得到一个新的表
- 需要注意的是，使用 `UNION ALL` 进行表拼接时，要求两段 `SELECT` 语句产生的小表的**列数、列名、列的数据类型**必须一致，否则会报错
- 除此之外，每段 `SELECT` 语句中还可以包含一些其他的 SQL 语句，比如 `WHERE`、`GROUP BY`、`ORDER BY` 等等，这些 SQL 语句会在每段 `SELECT` 语句中独立执行，不会相互影响，只要保证每段 `SELECT` 语句产生的小表的列数、列名、列的数据类型一致就可以了

除了 `UNION ALL` 以外，还有一个 `UNION` 关键字也可以用来进行表拼接，但是它和 `UNION ALL` 的区别在于：

- `UNION` 会在拼接完成后自动去重，也就是说，如果两段 `SELECT` 语句产生的小表中有重复的行，那么在使用 `UNION` 进行表拼接时，这些重复的行只会保留一行，其他的重复行会被丢弃掉
- `UNION ALL` 则不会去重，所有的行都会被保留，所以在使用 `UNION ALL` 进行表拼接时，如果两段 `SELECT` 语句产生的小表中有重复的行，那么在拼接完成后，这些重复的行都会被保留

我们在 Chinook 数据库中，使用 `UNION ALL` 和 `UNION` 来做一些练习：

- 查询 `InvoiceLine` 表中曾经被卖出过的 `TrackId`，以及 `Track` 表中 `Milliseconds` 大于 300000 的 `TrackId`，并将这两部分的结果拼接在一起，并且去重：

````python
%%sql
SELECT DISTINCT
    TrackId
FROM InvoiceLine

UNION

SELECT DISTINCT
    TrackId
FROM Track
WHERE Milliseconds > 300000
````

- 将 `Customer` 表中的 `FirstName` 和 `LastName` 列拼接成一个新的列 `FullName`，同样，在 `Employee` 表中也将 `FirstName` 和 `LastName` 列拼接成一个新的列 `FullName`，然后将这两部分的结果拼接在一起，将客户和员工的电话和 Email 信息都展示出来：

````python
%%sql
SELECT
    CONCAT(FirstName, ' ', LastName) AS FullName,
    Phone,
    Email
FROM Customer

UNION ALL

SELECT
    CONCAT(FirstName, ' ', LastName) AS FullName,
    Phone,
    Email
FROM Employee
````
