# Topic 1 - 数据库与 SQL 基础知识

本章将介绍关系型数据库的基本概念，以及 SQL 在关系型数据库中的作用和重要性，为后续章节的 SQL 查询学习打下坚实的基础。我们将通过实际案例来说明这些概念，并为同学们提供一个清晰的框架，帮助大家理解和应用 SQL 查询。

- **Topic 1.1 - 数据库与 SQL 语言简介**：了解数据库的基本概念，关系型数据库的特点，以及 SQL 语言的作用和重要性。
- **Topic 1.2 - 关系型数据库基本概念**：了解关系型数据库的核心概念，如表、行、列、主键、外键等，以及它们在数据库中的作用和关系。
- **Topic 1.3 - 本模块使用的数据库 - Chinook**：介绍本模块使用的 Chinook 数据库，包括其结构以及如何连接。
